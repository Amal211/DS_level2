```
# Single discrete variable

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('https://raw.githubusercontent.com/Amal211/python/main/StudentsPerformance.csv')

plt.hist(df['math score'])
plt.title('Distribution of math scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.show()
```

```
# Single discrete data using histogram in matplotlib library

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


df = pd.read_csv('https://raw.githubusercontent.com/Amal211/python/main/StudentsPerformance.csv')

# replace missing values in 'gender' culomn and convert it to numeric
df['gender'].fillna(df['gender'].mode()[0], inplace=True)
df['gender'] = pd.factorize(df.gender)[0]

plt.figure(figsize=(4, 4))
sns.barplot(x="gender", hue="gender", y="writing score", data=df)
plt.show()
```

```
# two discrete variables

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/amueller/advanced_training/master/data/adult.csv')
df['age-group'] = pd.cut(df.age, bins=[25,45,65,90], labels=['Young', 'Middle-aged', 'Old'])

#visualize the relationship between two discrete variables (the age and the work hours per week)
sns.barplot(x='age-group', hue='age-group', y='hours-per-week', data=df)
plt.title('Relationship between Age and work hours per week')
```

```
# single continuous variable

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


df = pd.read_csv('https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv')

sns.histplot(df['mpg'])
sns.set(style='darkgrid',)
plt.title('Distribution of miles per gallon of fuel')
```

```
# two continuous variables

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

array = np.array([['Ahmad',25,65.25,6,3.2],['Mazen',47,72.43,7,1.9],
                  ['Ali',38,71.55,5,2.4],['Salem',20,62.33,7,2.1],
                  ['Rami',33,85.43,8,1.4],['Alaa',30,97.35,7,1.7]])

df = pd.DataFrame(array, columns = ['Name','Age','Weight','Sleep-per-day', 'Water'])
df['Age'] = df['Age'].astype(int)
df['Weight'] = df['Weight'].astype(float)
df['Sleep-per-day'] = df['Sleep-per-day'].astype(int)
df['Water'] = df['Water'].astype(float)

sns.scatterplot(x = "Water", y = "Weight", data=df)
sns.set(style='darkgrid',)
plt.title('Relationship between Water consumed daily and a person Weight')
```

```
# discrete vs. continuous variables
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

array = np.array([[25,65.25,6,3.2],[47,72.43,7,1.9],[38,71.55,5,2.4],[20,62.33,7,2.1],[33,85.43,8,1.4],[30,97.35,7,1.7]])
df = pd.DataFrame(array, columns = ['Age','Weight','Sleep-per-day', 'Water'])
df['Age'] = df['Age'].astype(int)
df['Weight'] = df['Weight'].astype(float)
df['Sleep-per-day'] = df['Sleep-per-day'].astype(int)
df['Water'] = df['Water'].astype(float)

df_corr = df.corr()

sns.heatmap(df_corr)
plt.title('Relationship between daily sleep amount and a person weight')
```
