```
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph() #create a graph

G.add_nodes_from(['Ri', 'Dm', 'Hi', 'Ar', 'Haf', 'Makkah', 'Bur', 'Ska', 'Tab', 'Jiz', 'Jed', 'Naj', 'Abh', 'Yan', 'Medina', 'Bah']) #add multiple nodes

# adding edges

G.add_edge('Ri', 'Makkah')

G.add_edge('Dm', 'Ar')

G.add_edge('Dm', 'Haf')

G.add_edge('Dm', 'Ri')

G.add_edge('Hi', 'Ri')

G.add_edge('Hi', 'Bur')

G.add_edge('Hi', 'Medina')

G.add_edge('Haf', 'Ri')

G.add_edge('Makkah', 'Jed')

G.add_edge('Makkah', 'Jiz')

G.add_edge('Bur', 'Ri')

G.add_edge('Jed', 'Medina')

G.add_edge('Jed', 'Yan')

G.add_edge('Jed', 'Jiz')

G.add_edge('Jiz', 'Abh')

G.add_edge('Jiz', 'Naj')

G.add_edge('Bah', 'Abh')

G.add_edge('Abh', 'Naj')

G.add_edge('Tab', 'Ska')

G.add_edge('Ska', 'Ar')

G.add_edge('Ska', 'Hi')

G.add_edge('Medina', 'Yan')


nx.draw(G, with_labels=True, font_weight='bold', node_size=1200)

plt.show()

# BFS

graph = G
visited = [] # List to keep track of visited nodes.
queue = []     #Initialize a queue

def bfs(visited, graph, node):
  visited.append(node)
  queue.append(node)

  while queue:
    s = queue.pop(0) 
    print (s, end = " ") 

    for neighbour in graph[s]:
      if neighbour not in visited:
        visited.append(neighbour)
        queue.append(neighbour)

# Driver Code
bfs(visited, graph, 'Ri')


# This is my order for BFS
# Ri Makkah Bur Hi Haf Dm Ar Ska Tab Medina Jed Yan Jiz Abh Naj Bah
```
