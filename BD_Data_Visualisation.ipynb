- Data visualization is the presentation of data by using graphs and charts. We need to visualize data in order to make them clearer and easier to understand by the audience, especially non-technical ones. 
- Data visualization can help decision-makers get insight from data, discover trends/patterns, find correlations, or find outliers even in an extensive data set with multiple dimensions (Seif, 2018).

- In Python, there are two popular packages used in data visualization: Matplotlib and Seaborn. Matplotlib is used for creating basic graphs such as bars, lines, scatterplots, and histograms. Seaborn is built on Matplotlib, but it is considered as a superset of it. Unlike Matplotlib, it has lower syntax and supports making high-level statistical graphics with pandas and NumPy libraries (Misal, 2019).


```python
# load the data and show the first 5 records

import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Amal211/DS_level2/main/gld_price_data.csv')

df.head()
```

```python
# show data frame information
# It has 6 columns (5 are float, one is string) and 2290 entries

df.info()
```

```python
import datetime as dt
from datetime import date, datetime
import numpy as np

# convert the Date column from string to date type

df['Date'] = pd.to_datetime(df['Date'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# use hist () to plot the distribution of Gold prices

GLD = df.GLD

plt.hist(GLD, bins=50)
plt.title('Distribution of the Gold price')
plt.xlabel('Gold price')
plt.ylabel('Frequency')
plt.xlim(0, 220)
plt.ylim(0, 180)
plt.show()
```

```python
# use scatter () to plot the relation between Gold and Silver prices

x = df.GLD 
y = df.SLV
plt.scatter(x, y, linewidth=1.5)
plt.title('Relationship between Gold price and the Silver price')
plt.xlabel('Gold Price')
plt.ylabel('Silver price')
plt.xlim(0, 200)
plt.ylim(0, 50)
plt.show()
```

```python
# convert the date to ordinal number to use it in plot

print(date.toordinal(datetime(2008, 1, 1, 0, 0)))
```

```python
# use lineplot () to represent the Gold price from the year of 2008 to 2018

import seaborn as sns

sns.set(style="darkgrid")
plt.figure(figsize=(16,8))

x = df.Date
y = df.GLD

sns.lineplot(x, y, linewidth=1.5).set_title('GLD Price in ten years')
plt.xlabel('Date')
plt.ylabel('Gold price')
plt.xlim(733042, 736695)
plt.ylim(0, 180)
plt.show()
```
